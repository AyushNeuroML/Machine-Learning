```python
import pandas as pd

my_csv = pd.read_csv('/home/ayushz/Kaggle/Datasets/tested.csv')

df = pd.DataFrame(my_csv)
```

```python
print(df.head(3))
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
print(df['Age'])
```

```python
df['Cabin'].fillna('Unknown',inplace=True)
df['Deck'] = df['Cabin'].str[0] # will take first letter of cabin
```

```python


df["Deck"].head(50)
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output=False)

deckk_df = ohe.fit_transform(df[['Deck']])
deck_df = pd.DataFrame(deckk_df, columns=ohe.get_feature_names_out(['Deck']))
print(deck_df.head())

```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()



# df['Sex_encoded'] = le.fit_transform(df['Sex'])


print(df.head(5))
```

```python
df.drop('Sex',axis=1,inplace=True)
```

```python
df['Encoded_Embarked'] = le.fit_transform(df['Embarked'])
```

```python

df.drop('Embarked', axis=1, inplace=True)

```

```python
finaldf = pd.concat([df, deck_df], axis=1)
```

```python
finaldf.drop('Cabin',axis=1,inplace=True)
finaldf.drop('Deck',axis=1,inplace=True)
```
