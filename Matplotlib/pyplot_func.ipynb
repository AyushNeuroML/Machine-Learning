```python
# plt.plot(x,y, color='color_name', linestyle = 'line_style', linewidth=value , marked = 'marker_symbol', label = 'label_name')

import matplotlib.pyplot as plt

months = [1,2,3,4]
sales = [1000,1500,1200,1800]


plt.plot(months , sales ,color = 'blue', linestyle = '--',linewidth = 2, marker = 'o',label='2025 sales data')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Monthly Sales Data for 2025')
# plt.legend(loc='upper left',fontsize=12)
plt.legend(loc='lower right',fontsize=12)
plt.grid(color='red',linestyle=':',linewidth=2)
plt.ylim(0,2000)
plt.xlim(1,4)
# plt.ylim(0,1800)
# plt.xlim(1,3)
plt.xticks([1,2,3,4],['M1','M2','M3','M4'])
plt.show()
```
