```python
import matplotlib.pyplot as plt


# fig , ax = plt.subplot(nrows,ncols,figsize=width,height)

fig, ax = plt.subplots(1,2,figsize = (10,5))

x = [1,2,3,4]
y = [10,20,15,25]

ax[0].plot(x,y,color='blue')
ax[0].set_title('Line Plot')


ax[1].bar(x,y,color='green')
ax[1].set_title('Bar Char')

fig.suptitle('comparison of line and bar chart')

plt.tight_layout()

# saving the images or plots or charts or visualised studff whatever you want to call it
# always remember do plt.show() after the plt.savefig()

plt.savefig('plots.png',dpi=300,bbox_inches='tight')
plt.show()




```
