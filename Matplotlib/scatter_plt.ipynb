```python
import matplotlib.pyplot as plt

hours_studies = [1,2,3,4,5,6,7,8]
scores = [50,60,70,80,90,100,110,120]

plt.scatter(hours_studies, scores , color = 'red', marker = 'o', label = 'Student Data')
plt.title("Scatter Plot of Study Hours vs Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Scores")
plt.legend()
plt.grid(True)
plt.show()
```
